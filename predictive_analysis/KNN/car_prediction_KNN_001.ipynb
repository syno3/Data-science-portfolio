### importing required modules

```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.preprocessing import LabelEncoder
```

### upload dataset to googlecould

```python
from google.colab import files
uploaded = files.upload() 
```

### read data file and column

```python
data = pd.read_csv('car.data', names=["buying", "maint", "doors", "persons","lug_boot","safety", 'class'])
data
```

### mapping data for x and y

```python
#mapping x

x_data = data.drop('class', axis=1)
x = x_data.apply(LabelEncoder().fit_transform)

#mapping y

data['class']=data['class'].astype('category')
y = data['class'].cat.codes
```

### create model

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights='uniform')
```

### create a test size

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size =0.3)
```

### train model

```python
knn.fit(x_train, y_train)
```

### get prediction and accuracy

```python
prediction = knn.predict(x_test)
accuracy = metrics.accuracy_score(y_test, prediction)
```

### print out the prediction and accuracy

```python
print('The predictions: ', prediction)
print('The accuracy: ', accuracy)
```
